# Web Scrap -  EGX 30 Egypt Stock Index

###### The code should be run after stock working hours 

```python
## Import requests to handle http requests
## Import Beautifulsoup to parse html scripts

import requests
from bs4 import BeautifulSoup

url= 'https://english.mubasher.info/countries/eg'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
print(soup)
```

```python
data = soup.find('div', attrs={'class':'mi-indices-summary__content md-whiteframe-z1 mi-js-tab__content'}).text
print(data)
```

```python
print(data.split())
```

```python
## handling received data from string to integers and float
egx = data.split()
close = float(egx[1].replace(',',''))
change = float(egx[2].replace('%',''))
prev_close =  format((close - change), '.2f')
volume = int(egx[5].replace(',',''))
turnover = int(egx[7].replace(',',''))
trades_num = int(egx[11].replace(',',''))

print(close, change,prev_close, volume, turnover,trades_num)
```

```python
## Impoort sqlite3 to handle sql database and fill data in EGX_Stock.db
## also fill data in a text file

import sqlite3
from datetime import datetime, date

date = datetime.now().date()


stock = sqlite3.connect("EGX_Stock.db")
cr = stock.cursor()
cr.execute("create table if not exists users (date text, close integer, change integer, prev_close integer, volume integer, turnover integer, trades_num integer)")
cr.execute(f"insert into users (date, close, change, prev_close, volume, turnover, trades_num) values ('{date}', {close}, {change}, {prev_close}, {volume}, {turnover}, {trades_num})")
stock.commit()
stock.close()


f = open('egx_stock.txt', 'a')
output = f"date = {date} close= {close}, change= {change}, previous close= {prev_close}\nvolume = {volume}, turnover= {turnover}, trades_num= {trades_num}\n\n"
f.write(output)
f.close()
```
